```python
from bookinz.storage import BronzeLayer
```

```python
bl = BronzeLayer("data")

# Cheapest available hotels in Amsterdam
df = bl.query("""
    with query as (
        select
            facility_id,
            name,
            currency,
            search_area,
            neighbourhood,
            accommodation_type,
            tags,
            distance_from_center_km,
            date_diff('day', cast(checkin_date as date), cast(checkout_date as date)) as num_nights,
            total_price,
            round(total_price / nullif(date_diff('day', cast(checkin_date as date), cast(checkout_date as date)), 0), 2) as price_per_night,
            round((total_price) / num_adults, 2) as price_per_person,
            round((total_price) / nullif(date_diff('day', cast(checkin_date as date), cast(checkout_date as date)) * num_adults, 0), 2) as price_per_night_per_person,
            url
        from bronze
              
    )
    
    select
        *
    from query
    where
        1 = 1 
        and price_per_night_per_person <= 80
        --and (
        --      (distance_from_center_km <= 3 and search_area = 'Budapest__Hungary')
        --      or
        --      (distance_from_center_km <= 1.5 and search_area = 'Ponta_Delgada__Portugal')
        --    )

""")
df
```
