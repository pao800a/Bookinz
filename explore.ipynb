```python


# Cheapest available hotels in Amsterdam
df = bl.query("""
with 
    bronze_fetched as (
        select
            name, 
            search_area, 
            total_price, 
            date_diff('day', cast(checkin_date as date), cast(checkout_date as date)) as num_nights,
            round(total_price / date_diff('day', cast(checkin_date as date), cast(checkout_date as date)), 2) as price_per_night,
            round(total_price / num_adults, 2) as price_per_person,
            round(total_price / (date_diff('day', cast(checkin_date as date), cast(checkout_date as date)) * num_adults), 2) as price_per_night_per_person,
            rating, 
            round(rating / round(total_price / (date_diff('day', cast(checkin_date as date), cast(checkout_date as date)) * num_adults), 2), 2) as quality_score, 
            num_reviews, 
            distance_from_center_km, 
            num_rooms_available, 
            neighbourhood, 
            accommodation_type, 
            tags
        from bronze
        where search_area like '%Italy'
    )
    select
        name,
        search_area,
        total_price,
        price_per_night,
        price_per_person,
        price_per_night_per_person,
        rating,
        (quality_score - MIN(quality_score) OVER ()) / NULLIF(MAX(quality_score) OVER () - MIN(quality_score) OVER (), 0) AS quality_score,
        num_reviews,
        distance_from_center_km,
        num_rooms_available,
        neighbourhood,
        accommodation_type,
        tags     
    from bronze_fetched where quality_score is not null order by quality_score desc
              
"""
)
df
```
